#### 3. Create a connection to BigQuery

```python
client = bigquery.Client() 
```

#### 4. Import data from the users table from the stackoverflow public database for United Kingdom only

```python
query='''
SELECT
    *
FROM
    `bigquery-public-data.stackoverflow.users`
WHERE
    location LIKE '%UK%

'''

query_job = client.query(query=query)
df=query_job.to_dataframe()
df.head()
```

#### 5. Rename id column to user_id

```python
df = df.rename(columns={'id':'user_id'})
df.head()                   
```

#### 6. Import from the stackoverflow_posts table all items which have up to 20 view counts and only select 'id', 'owner_user_id', 'score', 'view_count', 'comment_count', 'favorite_count'

```python
query='''
SELECT
    id, 
    owner_user_id,
    score,
    view_count,
    comment_count,
    favorite_count
    
FROM
    `bigquery-public-data.stackoverflow.stackoverflow_posts`
WHERE
    view_count <= 20

'''

query_job = client.query(query=query)
df2=query_job.to_dataframe()
df2.head()
```

#### 7. Rename id column to post_id and owner_user_id to user_id

```python
df2 = df2.rename(columns={'id':'post_id', 'owner_user_id':'user_id'})
df2.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: user_id, reputation,views,up_votes,down_votes
    **posts columns**: post_id, score,user_id,view_count,comment_count

```python
users_df = df[['user_id', 'reputation', 'views', 'up_votes', 'down_votes']]
users_df.head()
```

```python
posts_df = df2[['post_id', 'score','user_id','view_count','comment_count']]
posts_df.head()
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_df = users_df.merge(posts_df, on='user_id')
merged_df.head()
```

#### 9. How many missing values do you have in your merged dataframe? 

```python
merged_df.isnull().sum()
#there are no missing values in the new dataframe
```

## Bonus: 


### Identify extreme values in your merged dataframe, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder. Hint: post_id cannot have outliers!

```python
stats = merged_df.describe().T
stats
```

```python
outliers = pd.DataFrame(columns=merged_df.columns)

#cols_lst contain the columns with possible outliers
cols_lst = ['reputation', 'views', 'up_votes', 'down_votes', 'score', 'view_count', 'comment_count']

for col in cols_lst:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    
    results = merged_df[(merged_df[col] < lower) | (merged_df[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
outliers.head()
```

```python
outliers.to_csv("C:/Users/marcm/git-files/lab-data-cleaning/your-code/outliers.csv")
```
