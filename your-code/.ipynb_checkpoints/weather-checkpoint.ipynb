## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
weather = pd.read_csv('../weather-raw.csv')
```

```python
weather.head()
```

```python
#cuento valores nulos en cada columna

null_per_col = weather.isnull().sum()
null_per_col[null_per_col>0]

#la mayoria de dias no tienen lectura de temperatura
```

```python
#hago melt de todas las columnas de los dias
weather_melt = weather.melt(id_vars=['id','year','month','element'],var_name= 'day', value_name='temp')
weather_melt
```

```python
#hago pivot_table (no pivot porque tengo valores repetidos y no lo permite) 
#separando lecturas de temp max y min en columnas

weather_pivot = weather_melt.pivot_table(index=['year','month','day'], columns='element')
weather_pivot

#los valores NaN han desaparecido (???)
```
