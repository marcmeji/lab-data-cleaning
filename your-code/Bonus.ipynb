```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:/Users\marcm\git-files\lab-data-cleaning\weather-raw.csv") 
data
```

```python
#cuento valores nulos en cada columna

null_per_col = data.isnull().sum()
null_per_col[null_per_col>0]

#la mayoria de dias no tienen lectura de temperatura
```

```python
#los valores id y year son siempre los mismos. Como estas columnas no aportan informacion, las elimino

data = data.drop(['id', 'year'], axis=1)
```

```python
data
```

```python
#cambio numero de mes por su nombre

data['month'] = data['month'].replace([1,2,3,4,5,6,7,8,10,11,12],['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Oct','Nov','Des'])
```

```python
data
```

```python
#nueva columna con el promedio de temperatura de cada fila

data['mean'] = round(data.mean(axis=1),2)
data
```

```python
#elimino lecturas diarias, puesto que la mayoria estan vacias

data = data.drop(['d1','d2','d3','d4','d5','d6','d7','d8','d9','d10','d11','d12','d13','d14','d15',
           'd16','d17','d18','d19','d20','d21','d22','d23','d24','d25','d26','d27','d28','d29',
           'd30','d31'],axis=1)
```

```python
#separo valores maximo y minimopara cada mes en dos columnas

data = data.pivot(index='month', columns='element', values='mean')
data
```
